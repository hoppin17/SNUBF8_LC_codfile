```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/model/eq1_best_threshold_sec.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/model/eq2_best_threshold_sec.csv')
```

```python
df1
```

```python
import matplotlib.pyplot as plt

col_list = list(df1.columns)

# 새로운 Figure 객체 생성
plt.figure(figsize=(18, 10))

# 6개의 서브플롯 생성
for num in range(6):
    plt.subplot(2, 3, num + 1)  # 2행 3열 그리드의 num+1번째 서브플롯
    plt.hist(df1[col_list[num+1]])
    plt.title(col_list[num+1])  # 제목을 열 이름으로 설정
    plt.xlabel('Value')  # x축 레이블 설정
    plt.ylabel('Frequency')  # y축 레이블 설정

# 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```
